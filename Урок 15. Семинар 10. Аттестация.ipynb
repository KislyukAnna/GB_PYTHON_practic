```python
import pandas as pd
import random

lst = ['robot'] * 10
lst += ['human'] * 10
random.shuffle(lst)
data = pd.DataFrame({'whoAmI':lst})
data.head()
```

```python
# 1 способ

data1 = pd.get_dummies(data, columns=['whoAmI'])
data1.head()
```

```python
# 2 способ

list2 = list(data['whoAmI'].unique())
# list2
for i in list2:
  data.loc[data['whoAmI'] == i, i] = 1
  data.loc[data['whoAmI'] != i, i] = 0
data.head()
```
